```python
import pandas as pd
import numpy as np

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)

rydde_data = pd.read_excel("rydde_data/Ryddestatistikk 17-03-2023.xlsx")
```

# The totals

General data cleaning up:

```python
rydde_data = rydde_data[rydde_data.Deltakere !=0]
rydde_data = rydde_data[rydde_data.Kg !=0]

rydde_data = rydde_data.drop(columns = ['Aksjonskode', 'AksjonsID', 'Aksjonsnavn',  'Offentlig',
       'Registrering fullført', 'Rydde app', 'Min bit av Norge'])
rydde_data['Start'], rydde_data['Slutt'] = pd.to_datetime(rydde_data['Start']), pd.to_datetime(rydde_data['Slutt'])
rydde_data = rydde_data.reset_index()
rydde_data = rydde_data.drop(columns = ['index'])
```

```python
rydde_data['Most found element'] = rydde_data.iloc[:, 12:-16].idxmax(axis = 1)
rydde_data.head()
rydde_data.to_csv('new_data/RyddeDataCleanedUp.csv')
```

```python
total_kg_waste = rydde_data['Kg'].sum()
total_involved_people = rydde_data['Deltakere'].sum()
total_m_cleaned = rydde_data['Meter'].sum()
total_bags_filled = rydde_data['Søppelsekker'].sum()

rydde_timeframes = pd.DataFrame({'Total_KG_Waste': [total_kg_waste], 'Total_involvement_people': [total_involved_people], 'Total_m_cleaned': [total_m_cleaned], \
                       'Total_bags': [total_bags_filled]})
days = np.abs((rydde_data['Start'] - rydde_data['Slutt']).astype('timedelta64[D]'))
print(days)
# rydde_timeframes['Days_duration']
days = days.replace(to_replace=0.0, value = 1)

rydde_timeframes['Total_Duration'] = days.sum()
```

```python
rydde_timeframes.to_csv('new_data/TotalValues.csv', index = False)
```
