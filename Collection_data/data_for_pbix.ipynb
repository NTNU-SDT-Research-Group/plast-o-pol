# Yearly Totals

Easy, general totals:

```python
# get rid of some data that are not useful here
rydde_data = rydde_data.drop(columns = ['id','name', 'start_hour','end_hour','organisation','startyear', 'endyear','description','cleaningteam_type',\
                                     'operation_type', 'operationmeetingplace','published', 'handout_gloves', 'handout_wastebags', \
                                        'cleaningteam', 'municipality', 'county',	'join_code', 'cleaning_code', \
                                            'registration_completed', 'adopted'])

total_kg_waste = rydde_data['kg_of_collected_waste'].sum()
total_involved_people = rydde_data['number_of_participants'].sum()
total_m_cleaned = rydde_data['meters_cleaned_beach'].sum()
total_bags_filled = rydde_data['number_of_waste_bags'].sum()
```

Now the totals but per county:

```python
# make new dataframe only with the totals we are looking for:
rydde_regional2 = rydde_data.sort_values(by=['county_name']).reset_index()
counties = rydde_regional2.pop('county_name')
rydde_regional2.insert(0,'county_name',counties)
# dff = df.group by(["SINID","EXTRA"]).MONTREGL.sum().reset_index()
rydde_regional = rydde_regional2.groupby(["county_name"]).sum().reset_index()
rydde_regional = rydde_regional.drop(rydde_regional.index[6]).reset_index()
rydde_regional = rydde_regional.drop(columns = ['index'])
rydde_regional = rydde_regional.drop(columns = ['level_0'])
```

# Temporal data
From the start-end date

```python
rydde_data['startdate']
```

```python
rydde_times = pd.DataFrame({'start_time': rydde_data['startdate'], 'end_time': rydde_data['enddate'], 'kg_plastic_waste': rydde_data['kg_of_collected_waste']})
# rydde_times['enddate'] = rydde_data['enddate']
# df['duration_hours'] = (df['endtime'] - df['starttime']).astype('timedelta64[h]')
rydde_times['duration_days'] = (rydde_times['end_time'] - rydde_times['start_time']).astype('timedelta64[D]')

# change all 0's in column_A to 1's (cause it was at least one day)
rydde_times.loc[rydde_times['duration_days'] == 0, 'duration_days'] = 1

rydde_times['duration_days']
```

```python
# convert the starttime column to datetime format
rydde_times['starttime'] = pd.to_datetime(rydde_times['start_time'])

# DAYS SPENT CLEANING
num_dates = rydde_times['duration_days'].sum()
num_dates
```

```python
rydde_times['month'] = rydde_times['start_time'].dt.month_name()
```

```python
rydde_times['start_time'].astype('timedelta64[Y]')
```

```python
rydde_times # point that start_time and end_time is a little messed up
```

```python
rydde_times = rydde_times.groupby(['month']).sum().reset_index()
```

```python
months = ["January","February", "March", "April", "May", "June", 
          "July", "August", "September", "October", "November", "December"]

rydde_times['month'] = pd.Categorical(rydde_times['month'], categories=months, ordered=True)

rydde_times = rydde_times.sort_values(by='month').reset_index()

rydde_times = rydde_times.drop(columns=['index'])

# df['CUMSUM_C'] = df['SUM_C'].cumsum()

rydde_times['CUMSUM'] = rydde_times['kg_plastic_waste'].cumsum()
rydde_times
```
