# Part 1. Making data homogeneous

On the data that I chose to use (since I had many different options and received data at different points in life).
```python
import pandas as pd
import numpy as np

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)

rydde_data2020 = pd.read_excel("rydde_data/2-Rydde - Aksjoner med funn _Ny protokoll_.xlsx")
rydde_data2021 = pd.read_excel("rydde_data/Aksjoner med funn _Ny protokoll_fra april 2020, med siste data fra nylige ryddeaksjoner.xlsx")
rydde_data2023 = pd.read_excel('rydde_data/Kopi av Rydde - Aksjoner med funn _Ny protokoll_2022 og 2023 pr 160623.xlsx')
```

Checking the years that are available in the data:

Cleaning up the data

```python
# For rydde 2022/23
rydde_data2023 = rydde_data2023[rydde_data2023.Meter !=0]
rydde_data2023 = rydde_data2023[rydde_data2023.Kg !=0]

rydde_data2023 = rydde_data2023.drop(columns = ['System id', 'Offentlig', 'Aksjonsnavn', 'Organisasjon','Beskrivelse', 'Ryddelag',
                                                    'Type ryddelag', 'Aksjonstype', 'Møteplass', 'Publisert', 'Hansker utlevers','Sekker utleveres',
                                                     'Kommunenummer', 'Fylkesnummer', 'Påmeldingskode', 'Aksjonskode', 'Antall bilder','Aksjonsstatus',
                                                      'geom_wkt', 'Antall minutter ryddet', 'Lavterskel aksjoner', 'Adoptert',  'Registrert funn'])

rydde_data2023['Startdato'], rydde_data2023['Sluttdato'] = pd.to_datetime(rydde_data2023['Startdato']), pd.to_datetime(rydde_data2023['Sluttdato'])
rydde_data2023 = rydde_data2023.reset_index()
rydde_data2023 = rydde_data2023.drop(columns = ['index'])
```

```python
# for rydde 2020
rydde_data2020 = rydde_data2020[rydde_data2020.number_of_participants !=0]
rydde_data2020 = rydde_data2020[rydde_data2020.kg_of_collected_waste !=0]

rydde_data2020 = rydde_data2020.drop(columns = ['id', 'name', 'organisation', 'serveringssteder', 'annen_arsak', 'kommentar', 'media_files',
       'geom_wkt',  'join_code', 'cleaning_code', 'description', 'cleaningteam',
       'cleaningteam_type', 'operation_type', 'operationmeetingplace',
       'published', 'adopted','registration_completed', 'dumpet_avfall', 'havn', 'county', 
       'avfallsstasjon', 'bygg_anlegg', 'fiskeri_havbruk', 'handout_gloves', 'handout_wastebags','kloakkutslipp','municipality'])

rydde_data2020['startdate'], rydde_data2020['enddate'] = pd.to_datetime(rydde_data2020['startdate']), pd.to_datetime(rydde_data2020['enddate'])
rydde_data2020 = rydde_data2020.reset_index()
rydde_data2020 = rydde_data2020.drop(columns = ['index'])
```

```python
# for rydde 2021
rydde_data2021 = rydde_data2021[rydde_data2021.number_of_participants !=0]
rydde_data2021 = rydde_data2021[rydde_data2021.kg_of_collected_waste !=0]

rydde_data2021 = rydde_data2021.drop(columns = ['id', 'name', 'organisation', 'serveringssteder', 'annen_arsak', 'kommentar', 'media_files',
       'geom_wkt',  'join_code', 'cleaning_code', 'description', 'cleaningteam',
       'cleaningteam_type', 'operation_type', 'operationmeetingplace','public',
       'published', 'adopted','registration_completed', 'dumpet_avfall', 'havn', 'county',
       'avfallsstasjon', 'bygg_anlegg', 'fiskeri_havbruk', 'handout_gloves', 'handout_wastebags','kloakkutslipp','municipality'])

rydde_data2021['startdate'], rydde_data2021['enddate'] = pd.to_datetime(rydde_data2021['startdate']), pd.to_datetime(rydde_data2021['enddate'])
rydde_data2021 = rydde_data2021.reset_index()
rydde_data2021 = rydde_data2021.drop(columns = ['index'])
```

Some columns that are observed different in the new version of data from Rydde can already be changed below.

```python
rydde_data2023.rename(columns = {'Startdato':'startdate', 
                                 'Sluttdato':'enddate', 
                                 'Antall Deltakere':'number_of_participants',
                                 'Antall Søppelsekker':'number_of_waste_bags', 
                                 'Kg':'kg_of_collected_waste',
                                 'Fylkesnavn':'county_name',
                                 'Kommunenavn':'municipality_name',
                                 'Turisme': 'turisme',
                                 'Startår':'startyear', 
                                 'Sluttår': 'endyear',
                                 'Starter': 'start_hour',
                                 'Slutter': 'end_hour',
                                 'Meter': 'meters_cleaned_beach',
                                 'Rør':'ror',
                                 'Oljefat':'oljefat',
                                 'Antall deltakere': 'number_of_participants',
                                 'Metallbokser – utenlandske': 'metallbokser_vet_ikke'
                                 }, inplace = True)

rydde_data2023['number_of_waste_bags'] = rydde_data2023['Antall søppelsekker'] + rydde_data2023['Antall handleposer']
rydde_data2023 = rydde_data2023.drop(columns = ['Registrering fullført','Avfallsstasjon','Kloakkutslipp',
                                                'Havn','Dumpet avfall','Fiskeri og havbruk','Bygg og anlegg','Serveringssteder','Annen årsak',
                                                'Kommentar', 'Antall handleposer', 'Antall søppelsekker'])
```

Comparing all the dataframes (2020,2021,2023) to see what the differences are. I put the column names on an Excel that I then used to rename all the columns from one standard to the other (2023 was renamed to fit 2020 and 2021)


```python
 # compare rydde 2020 columns with rydde 2021
columns_missingfrom2021 = set(rydde_data2020.columns) - set(rydde_data2021.columns) 
columns_missingfrom2020 = set(rydde_data2021.columns) - set(rydde_data2020.columns)


columns_missingfrom2023 = set(rydde_data2020.columns) - set(rydde_data2023.columns)

assert columns_missingfrom2020 == columns_missingfrom2021, "2020 and 2021 datasets are the same"

print('Names for the 2023 dataset are in Norwegian. Here the differences \n', columns_missingfrom2023 )
```

Introduce the mapping dataset , an Excel file where the column names were mapped one by one.

```python
mapping_file = 'ColumnsToChangeV1.xlsx'
mapping_order_df = pd.read_excel(mapping_file)

for row in mapping_order_df.itertuples(index=False):
    column_name_2023 = row._1
    column_name_2023_new = row._0
    rydde_data2023.rename(columns={column_name_2023: column_name_2023_new}, inplace=True)
```

```python
columns_missingfrom2023 = set(rydde_data2023.columns) - set(rydde_data2020.columns)

print('Names for the 2023 dataset are in Norwegian. Here the differences \n', columns_missingfrom2023)
```

Had to repeat the process because it seems that I had some obsolete names? Or that there were some duplicates. This time the names are saved in the second worksheet.

```python
mapping_order_df2 = pd.read_excel(mapping_file, sheet_name='Additional Column Names')

for newrow in mapping_order_df2.itertuples(index=False):
    column_name_2023_2 = newrow._1
    column_name_2023_2_new = newrow._0
    rydde_data2023.rename(columns={column_name_2023_2: column_name_2023_2_new}, inplace=True)
```

```python
columns_missingfrom2023 = set(rydde_data2023.columns) - set(rydde_data2020.columns)

print('Names for the 2023 dataset are in Norwegian. Here the differences \n', columns_missingfrom2023)
```

# Part 2: Make data for PowerBI

Where the separate sub-excels are made so they can be loaded into PowerBi directly.

```python
# let's make one single dataframe
rydde = pd.concat([rydde_data2020,rydde_data2021,rydde_data2023], sort=False,ignore_index=True)

rydde = rydde.fillna(0)
rydde = rydde.drop(columns='annet')

rydde.head()
```

These are the cumulative data for the month and for the totals

```python
rydde_monthly =  rydde.copy()
rydde_monthly['month'] = rydde_monthly['startdate'].dt.to_period('M')
rydde_monthly = rydde_monthly.groupby(['month']).sum().reset_index()
rydde_monthly['CUMSUM'] = rydde_monthly['kg_of_collected_waste'].cumsum()
rydde_monthly_final = pd.DataFrame(data={'cumsum':rydde_monthly['CUMSUM'], 'month':rydde_monthly['month']})
```

```python
rydde_monthly.to_csv('processed_data/NEW_MonthCumSum.csv', index=False)
```

Most found in the country:

```python
rydde_municipality = rydde.copy()
rydde_municipality = rydde_municipality.groupby(["county_name"]).sum().reset_index()

rydde_municipality.head()
```

```python
rydde_municipality.to_csv('processed_data/NEW_County.csv')
```

The totals:

```python
total_involved_people = rydde['number_of_participants'].sum()
total_bags_filled = rydde['number_of_waste_bags'].sum()
total_kg = rydde['kg_of_collected_waste'].sum()

rydde_totals = pd.DataFrame(data={'People': [total_involved_people], 'Bags':[total_bags_filled], 'Kg': [total_kg]})

rydde_totals.to_csv('processed_data/NEW_Totals.csv')
```
